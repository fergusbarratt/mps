```python
%matplotlib inline
from spin import spins, N_body_spins
import numpy as np
from numpy.linalg import eigvals
import matplotlib.pyplot as plt
```

```python
Sx1, Sy1, Sz1 = N_body_spins(0.5, 1, 2)
Sx2, Sy2, Sz2 = N_body_spins(0.5, 2, 2)
n_z = Sz1-Sz2
def H(s, h): return 4*(Sx1@Sx2+Sy1@Sy2+Sz1@Sz2) - 1j*(s+1j*h)*n_z
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 5))
for h in np.linspace(0, 0.5, 5):
    S = np.linspace(-3, 3, 500)
    ax.plot(S, [max(eigvals(H(s, h)).imag) for s in S], label = '{:.1f}'.format(h))
plt.legend();
plt.savefig('images/dyn.pdf')
```
